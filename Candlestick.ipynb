### Candlestick

__Price__

```python
from pyecharts import options as opts
from pyecharts.charts import Kline
from pyecharts.charts import Line
```

```python
len(quarterly)
```

```python
ma = quarterly.close.rolling(window=10).mean().round(2).tolist()
```

```python
def kline() -> Kline:
    kline = (
        Kline()
        .add_xaxis(quarterly.index.strftime('%F').tolist())
        .add_yaxis(
            'Kline',
            quarterly.values.tolist(),
            itemstyle_opts=opts.ItemStyleOpts(
                color='#DD2200',
                color0='#009933',
                border_color='#DD2200',
                border_color0='#009933',
            ),
        )
        .set_global_opts(
            xaxis_opts=opts.AxisOpts(is_scale=True),
            yaxis_opts=opts.AxisOpts(
                is_scale=True,
                splitarea_opts=opts.SplitAreaOpts(
                    is_show=True, areastyle_opts=opts.AreaStyleOpts(opacity=1)
                ),
            ),
            # Set Legend
            legend_opts=opts.LegendOpts(pos_right=90),
            # Set DataZoom
            datazoom_opts=[opts.DataZoomOpts(type_='inside', range_start=40, range_end=100), opts.DataZoomOpts()],
            # Set Title
            title_opts=opts.TitleOpts(title='格力电器-000651.SZ', pos_left=80),
            # Set Tooltip 
            tooltip_opts=opts.TooltipOpts(
                trigger='axis',
                axis_pointer_type='cross',
                background_color='rgba(245, 245, 245, 0.8)',
                border_width=1,
                border_color='#ccc',
                textstyle_opts=opts.TextStyleOpts(color='#000'),
            ),
            # Set Axispointer
            axispointer_opts=opts.AxisPointerOpts(
                is_show=True,
                link=[{'xAxisIndex': 'all'}],
                label=opts.LabelOpts(background_color='#777'),
            )
        )
    )
    MAline = (
        Line()
        .add_xaxis(quarterly.index.strftime('%F').tolist())
        .add_yaxis(
            series_name='MA10',
            y_axis=ma,
            itemstyle_opts=opts.ItemStyleOpts(color='#2F4F4F'),
            is_symbol_show=False,
            is_smooth=True,
            is_hover_animation=False,
            linestyle_opts=opts.LineStyleOpts(width=2, opacity=0.7),
            label_opts=opts.LabelOpts(is_show=False),
        )
        .set_global_opts(xaxis_opts=opts.AxisOpts(type_='category'))
    )
    return kline.overlap(MAline)
```

```python
chart = kline()
chart.render_notebook()
```
